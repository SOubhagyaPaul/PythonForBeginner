```python
# Load the csv dataset “Netflix TV Shows and Movies” into a DataFrame called ratings.
import pandas as pd
file ='Netflix TV Shows and Movies.csv'
ratings=pd.read_csv(file)
ratings.head()
```

```python
# Select the description column from ratings and assign the result to the variable
# desc. What type of object is desc?
import pandas as pd
desc = ratings['description']
type(desc)
```

```python
# Select the first value from the description column of ratings, assigning it to variable
# first_description.
first_description=ratings['description'].iloc[0]
first_description
```

```python
# Select the first record from ratings, assigning it to the variable first_row.
first_row=ratings.iloc[0]
first_row
```

```python
# Select the first 15 values from the description column in ratings, assigning the result
# to variable first_descriptions.
first_descriptions=ratings['description'].iloc[:15]
first_descriptions
```

```python
# Select the records with index labels 1, 2, 4, 7, and 11, assigning the result to the variable
# sample_ratings.
sample_ratings=ratings.loc[[1,2,4,7,11]]
sample_ratings
```

```python
# Create a DataFrame df containing the title, type, release_year, and
# age_certification, columns of the records with the index labels 0, 4, 50, and 200
ind=[0,4,50,200]
col=['title', 'type', 'release_year','age_certification']
df=ratings.loc[ind,col]
df
```

```python
# Create a DataFrame df containing the title and release_year columns of the first 50
# records

sec_col=['title','release_year']
df=ratings.loc[:50][sec_col]
df
```

```python
# Create a DataFrame show_ratings containing ratings of shows. 
show=ratings[ratings['type']=="SHOW"]

df=show[['title','imdb_score']].iloc[:]
df
```

```python
#Create a DataFrame great_shows containing shows with ratings more than 7.5.
great_shows=df[df['imdb_score']>7.5]
great_shows
```

```python
# Create a DataFrame mini containing movies with runtime less than an hour.
mini=ratings[ratings['runtime']<60]
df=mini['title'].iloc[:]
df
```

```python
# Create a DataFrame my_age containing movies that were released in your year of birth. 
my_age=ratings[ratings['release_year']==2002]
df=my_age.loc[:][['title']]
df
```

```python
# Create a DataFrame my_movies containing movies that were released in the year 2000,
# 2004, 2008 and 20012. 
yr=[2000,2004,2008,2012]
my_movies=ratings[ratings['release_year'].isin(yr)]
my_movies
```

```python
# What is the mean, median and mode of the imdb_score column in the ratings
# DataFrame?
df=pd.DataFrame(ratings['imdb_score'])
print("mean: ",df.mean())
print("mode: ",df.mode())
print("median: ",df.median())
```

```python
# What release years are represented in the dataset (must not include duplicates)?
df = pd.DataFrame(ratings)
df.release_year.unique()
```

```python
# How often does each release year appear in the dataset?
df = pd.DataFrame(ratings)
df['release_year'].value_counts()
```

```python
#Which are the top three mostly voted movies
top_three_movies = ratings.nlargest(3, 'imdb_votes')
print(top_three_movies[['title', 'imdb_votes']])
```

```python
# What are the most common runtime of the movies in the dataset?
df['runtime'].value_counts().nlargest(3)
```

```python
# What are the highest rated and least rated movies yearwise?
view=ratings[ratings['imdb_score']==df['imdb_score'].max()]
view
```

```python
view=ratings[ratings['imdb_score']==df['imdb_score'].min()]
view
```

```python
# What combinations of runtime and rating are most common?
most_comm_combi = df.groupby(['runtime', 'imdb_score']).size().reset_index(name='count').sort_values(by='count',ascending=False).head(10)
most_comm_combi
```
