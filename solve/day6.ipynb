```python
# The popularity of programming languages as given by TIOBE index for November 2023:
# Programming language: Python, C, C++, Java, C#, JavaScript, PHP, Visual Basic
# Ratings: 14.16, 11.77, 10.36, 8.35, 7.65, 3.21, 2.30, 2.10
# For the given data, create
# (i) a bar chart.
# (ii) a horizontal bar chart.
# (iii) a pie chart.

import numpy as np
import matplotlib.pyplot as plt 
 
  
# creating the dataset
data = {'Python':14.16, 'C':11.77,'C++':10.36, 'Java':8.35, 
        'c#':7.65,'Javascript':3.21,'PHP':2.30,'visual basic':2.10}
courses = list(data.keys())
values = list(data.values())
  
#fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(courses, values)
 
plt.xlabel("Programming language")
plt.ylabel("Ratings")
plt.title("The popularity of programming languages as given by TIOBE index for November 2023")
plt.show()
print()
print("------a horizontal bar chart-----")
plt.barh(courses, values)
plt.xlabel("Programming language")
plt.ylabel("Ratings")
plt.title("The popularity of programming languages as given by TIOBE index for November 2023")
plt.show()

print()
print("------a horizontal bar chart-----")
plt.pie(values,labels=courses)
plt.show()
```
